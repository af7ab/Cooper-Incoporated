### Random Forest vs Logistic Regression Comparison

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Generate synthetic data
np.random.seed(42)
n_samples = 1000

# Features
X = pd.DataFrame({
    'Tackles': np.random.randint(0, 30, n_samples),
    'Interceptions': np.random.randint(0, 25, n_samples),
    'Blocks': np.random.randint(0, 20, n_samples),
    'Assists': np.random.randint(0, 15, n_samples),
    'Goals': np.random.randint(0, 20, n_samples),
    'xG': np.random.uniform(0, 2, n_samples),
    'xAG': np.random.uniform(0, 2, n_samples),
    'PassCompletion': np.random.uniform(70, 100, n_samples),
    'Stamina': np.random.uniform(50, 100, n_samples),
    'Speed': np.random.uniform(20, 40, n_samples)
})

# Target variable
y = np.random.choice(['Win', 'Loss/Draw'], size=n_samples, p=[0.6, 0.4])
y_binary = np.where(y == 'Win', 1, 0)

# Train-Test split
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

# Logistic Regression
log_model = LogisticRegression(random_state=42, max_iter=1000)
log_model.fit(X_train, y_train)
log_pred = log_model.predict(X_test)

# Metrics calculation
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

rf_scores = [
    accuracy_score(y_test, rf_pred),
    precision_score(y_test, rf_pred),
    recall_score(y_test, rf_pred),
    f1_score(y_test, rf_pred)
]

log_scores = [
    accuracy_score(y_test, log_pred),
    precision_score(y_test, log_pred),
    recall_score(y_test, log_pred),
    f1_score(y_test, log_pred)
]

# Plotting the comparison chart
x = np.arange(len(metrics))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 6))
bars1 = ax.bar(x - width/2, rf_scores, width, label='Random Forest', color='skyblue')
bars2 = ax.bar(x + width/2, log_scores, width, label='Logistic Regression', color='salmon')

# Adding text annotations
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

# Add labels, title, and legend
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Model Performance Comparison: Random Forest vs Logistic Regression')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

### Random Forest vs Logistic Regression Comparison

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# Generate synthetic data
np.random.seed(42)
n_samples = 1000

# Features from previous charts
X = pd.DataFrame({
    'Tackles': np.random.randint(0, 30, n_samples),
    'Interceptions': np.random.randint(0, 25, n_samples),
    'Blocks': np.random.randint(0, 20, n_samples),
    'Assists': np.random.randint(0, 15, n_samples),
    'Goals': np.random.randint(0, 20, n_samples),
    'xG': np.random.uniform(0, 2, n_samples),  # Expected goals
    'xAG': np.random.uniform(0, 2, n_samples),  # Expected assists
    'PassCompletion': np.random.uniform(70, 100, n_samples),  # Pass completion %
    'Stamina': np.random.uniform(50, 100, n_samples),  # Stamina rating
    'Speed': np.random.uniform(20, 40, n_samples)  # Speed rating
})

# Target variable
y = np.random.choice(['Win', 'Loss/Draw'], size=n_samples, p=[0.6, 0.4])

# Binarize the target for classification
y_binary = np.where(y == 'Win', 1, 0)  # 1 for Win, 0 for Loss/Draw

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Train Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision_win = precision_score(y_test, y_pred, pos_label=1)
precision_loss_draw = precision_score(y_test, y_pred, pos_label=0)
recall_win = recall_score(y_test, y_pred, pos_label=1)
recall_loss_draw = recall_score(y_test, y_pred, pos_label=0)
f1_score_win = f1_score(y_test, y_pred, pos_label=1)
f1_score_loss_draw = f1_score(y_test, y_pred, pos_label=0)

# Metrics for the bar chart
metrics = ['Accuracy', 'Precision (Loss/Draw)', 'Precision (Win)', 'Recall (Loss/Draw)', 'Recall (Win)', 'F1-Score (Loss/Draw)', 'F1-Score (Win)']
values = [accuracy, precision_loss_draw, precision_win, recall_loss_draw, recall_win, f1_score_loss_draw, f1_score_win]

# Plotting the chart
plt.figure(figsize=(10, 6))
bars = plt.bar(metrics, values, color='blue')

# Annotating the bars with values
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 0.05, f"{bar.get_height():.2f}", ha='center', color='white')

# Add labels and title
plt.title("Random Forest Model Performance")
plt.ylabel("Performance Value")
plt.xlabel("Metrics")
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)
plt.tight_layout()

# Show the plot
plt.show()
```
